```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
dias_map = {
    "Monday": "Segunda",
    "Tuesday": "Terça",
    "Wednesday": "Quarta",
    "Thursday": "Quinta",
    "Friday": "Sexta",
    "Saturday": "Sábado",
    "Sunday": "Domingo"
}
```

```python
xls = pd.read_excel("volumetria.xlsx", sheet_name=None, header=1)
xls["Agosto 2024"] = pd.read_excel("volumetria.xlsx", sheet_name="Agosto 2024", header=0)
sheets_to_use = list(xls.values())[:-4]
volumetria = pd.concat(sheets_to_use, ignore_index=True)
volumetria = volumetria.drop(columns="TOTAL.1")
```

## Tratamento de dados - Volumetria

```python
# Renomeando Colunas
volumetria.columns = ["Data", "RU_A", "RA_A", "RS_A", "Total_A",
                      "Data_2", "RU_J", "RA_J", "RS_J", "Total_J"]
volumetria = volumetria.drop(columns=["Data_2"])
volumetria = volumetria.fillna(0)
volumetria["Total_A"] = volumetria["RU_A"] + volumetria["RA_A"] + volumetria["RS_A"]
volumetria["Total_J"] = volumetria["RU_J"] + volumetria["RA_J"] + volumetria["RS_J"]
volumetria["Total_Dia"] = volumetria["Total_A"] + volumetria["Total_J"]
volumetria.head()
```

```python
# Obtendo dia da semana
volumetria["Data"] = pd.to_datetime(volumetria["Data"])
volumetria["Dia_Semana"] = volumetria["Data"].dt.day_name()
volumetria["Dia_Semana"] = volumetria["Dia_Semana"].map(dias_map)

volumetria["Mes"] = volumetria["Data"].dt.month
volumetria["Dia"] = volumetria["Data"].dt.day
volumetria["Ano"] = volumetria["Data"].dt.year

volumetria.head()
```

```python
df_final = pd.merge(
    volumetria,
    cardapio,
    left_on="Data",   # chave do volumetria
    right_on="date",  # chave do cardapio
    how="left"        # mantém todas as linhas do volumetria
)
```
